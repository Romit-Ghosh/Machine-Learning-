# OUTLIER DETECTION AND HANDLE

```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=df[['Aroma', 'Flavor', 'Aftertaste', 'Uniformity', 'Sweetness', 'Moisture', 'Quakers',
                     'Category.One.Defects', 'Category.Two.Defects']])
plt.title('Box Plot for Outlier Detection', fontsize=16, weight='bold')
plt.xlabel('Features', fontsize=14, weight='bold')
plt.ylabel('Values', fontsize=14, weight='bold')
plt.show()
```

```python
def detect_outliers_iqr(data):
    outlier_list = []
    data = sorted(data)
    
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    
    #print("The Val of Q1 and Q2",q1, q3)
    IQR = q3-q1
    
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    
    #print("The lower & Upper Bound",lwr_bound, upr_bound)
    
    for i in data: 
        if (i<lwr_bound or i>upr_bound):
            outlier_list.append(i)
    return outlier_list 



for i in features_num:
    outliers = detect_outliers_iqr(df[i])
    print("Outliers in",i,"attribute :", outliers)
```

```python
def handle_outliers(data):

    tenth_percentile = np.percentile(data, 10)
    ninetieth_percentile = np.percentile(data, 90)

    b = np.where(data<tenth_percentile, tenth_percentile, data)

    b1 = np.where(b>ninetieth_percentile, ninetieth_percentile, b)
    return b1
    
    
for i in features_num:
    df[i]=handle_outliers(df[i])
```
