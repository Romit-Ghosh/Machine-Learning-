# Model Training

```python
# Prepare the data for modeling
X = data.drop(columns='PurchaseStatus')
y = data['PurchaseStatus']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Standardize the continuous features
scaler = StandardScaler()
X_train[['Age', 'AnnualIncome', 'TimeSpentOnWebsite']] = scaler.fit_transform(X_train[['Age', 'AnnualIncome', 'TimeSpentOnWebsite']])
X_test[['Age', 'AnnualIncome', 'TimeSpentOnWebsite']] = scaler.transform(X_test[['Age', 'AnnualIncome', 'TimeSpentOnWebsite']])
```

```python
# Logistic Regression Model
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression(random_state=42)
logistic_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score
from sklearn import metrics
from sklearn.model_selection import GridSearchCV
```

```python
# Predictions and evaluation for Logistic Regression
y_pred_logistic = logistic_model.predict(X_test)
print("Logistic Regression:")
print(classification_report(y_test, y_pred_logistic))
print("Accuracy:", accuracy_score(y_test, y_pred_logistic))
```

```python
metrics.confusion_matrix(y_test, y_pred_logistic)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Decision Tree Model
tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train, y_train)
```

```python
# Predictions and evaluation for Decision Tree
y_pred_test = tree_model.predict(X_test)
print("Decision Tree:")
print(classification_report(y_test, y_pred_test))
print("Accuracy:", accuracy_score(y_test, y_pred_test))
```

```python
# Hyperparameter Tuning for Decision Tree
param_grid = {
    'criterion' : ['gini', 'entropy'],
    'max_depth': [3, 5, 7, 10, 12, 15, 16],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
grid_search = GridSearchCV(estimator=tree_model, param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
```

```python
# Best parameters and best score
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```

```python
# Evaluate tuned model
best_tree_model = grid_search.best_estimator_
y_pred_best_tree = best_tree_model.predict(X_test)
print("Tuned Decision Tree:")
print(classification_report(y_test, y_pred_best_tree))
print("Accuracy:", accuracy_score(y_test, y_pred_best_tree))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# K-Nearest Neighbors Model
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
```

```python
# Predictions and evaluation for KNN
y_pred_knn = knn_model.predict(X_test)
print("K-Nearest Neighbors:")
print(classification_report(y_test, y_pred_knn))
print("Accuracy:", accuracy_score(y_test, y_pred_knn))
```
